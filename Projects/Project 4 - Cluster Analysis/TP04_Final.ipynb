Se necesitan programas de socialización y capacitación completos para mejorar la facilidad de uso percibida y la adopción de los sistemas de información, especialmente para los estudiantes que requieren más apoyo.

En conclusión, el estudio enfatiza la necesidad de una socialización integral, un diseño de sistemas centrado en el usuario y manuales accesibles para mejorar la implementación de la tecnología de la información en la educación superior y mejorar la cultura digital.

### **iii. Marco teórico (que incluya la selección de los algoritmos a utilizar), definición del problema, el objetivo y las fuentes de datos que esperan utilizar.**

En el contexto de este proyecto de investigación, se han seleccionado los siguientes algoritmos de clustering:

**Affinity Propagation:** Este algoritmo es adecuado para identificar grupos o clusters en conjuntos de datos, sin la necesidad de especificar previamente el número de clusters esperados. Utiliza mensajes de afinidad entre pares de datos para propagar la información y encontrar los ejemplares más representativos del conjunto de datos. Su flexibilidad y capacidad para descubrir automáticamente la estructura subyacente lo convierten en una opción valiosa para nuestro análisis.

**Agglomerative Clustering:** Este algoritmo sigue un enfoque jerárquico, fusionando iterativamente los puntos más cercanos o grupos en función de una medida de distancia. Se inicia considerando cada punto como un grupo individual y luego fusiona los grupos más cercanos hasta obtener un único grupo. El número de clusters se determina mediante técnicas de corte o mediante la interpretación visual del dendrograma resultante.

**DBSCAN (Density-Based Spatial Clustering of Applications with Noise):** Este algoritmo se basa en la densidad de los puntos en el espacio de características. Identifica regiones densas de puntos como clusters y puede encontrar clusters de forma irregular y de diferentes tamaños. Además, puede detectar puntos que no pertenecen a ningún cluster, conocidos como ruido. DBSCAN es especialmente útil cuando los datos no están bien separados o cuando se espera encontrar clusters de densidades diferentes.

**KMeans:** Este es uno de los algoritmos de clustering más populares. Divide el conjunto de datos en k clusters, donde k se especifica previamente. El algoritmo asigna puntos al cluster más cercano en función de las distancias euclidianas y ajusta iterativamente los centroides de cada cluster para minimizar la suma de las distancias al cuadrado dentro del cluster. KMeans es rápido y eficiente en conjuntos de datos de grandes dimensiones.

# Affinity Propagation

## Affinity Propagation Clustering para los indicadores
* School enrollment, secondary
* Current education expenditure, secondary

```python
import numpy as np
from sklearn.cluster import AffinityPropagation
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt

# Filtered data for the selected indicators
indicator1 = 'SE.SEC.NENR'
indicator2 = 'SE.XPD.CSEC.ZS'
data = filtered_df.loc[filtered_df['Indicator Code'].isin([indicator1, indicator2])]

# Create data matrix X
X = np.array(data.pivot(index='Year', columns='Indicator Code', values='Value'))

# Perform imputation to handle missing values
imputer = SimpleImputer(strategy='mean')
X = imputer.fit_transform(X)

# Define the model
model = AffinityPropagation(damping=0.9)

# Fit the model
model.fit(X)

# Assign a cluster to each example
yhat = model.predict(X)

# Print the information for each data point
print("Data points:")
for i, cluster_label in enumerate(yhat):
    print(f"Data point {i+1}: Cluster {cluster_label}")

# Create scatter plot for samples from each cluster
clusters = np.unique(yhat)
for cluster in clusters:
    # Get row indexes for samples with this cluster
    row_ix = np.where(yhat == cluster)
    # Create scatter plot of these samples
    plt.scatter(X[row_ix, 0], X[row_ix, 1])

# Set labels and title
plt.xlabel('School enrollment, secondary (% net)')
plt.ylabel('Current education expenditure, secondary')
plt.title('Affinity Propagation Clustering')

# Show the plot
plt.show()
```

Los resultados del algoritmo de Affinity Propagation para los indicadores "SE.SEC.NENR: School enrollment, secondary (% net)" y "SE.XPD.CSEC.ZS: Current education expenditure, secondary (% of total expenditure in secondary public institutions)", podemos observar la asignación de clusters a cada uno de los puntos de datos.

Los resultados muestran que la mayoría de los puntos de datos (Data points 2 al 19) se han asignado al Cluster 1. Esto indica que estos puntos de datos comparten características similares en términos de los indicadores seleccionados. En particular, estos puntos de datos están asociados con una matriculación en educación secundaria (% neto) relativamente alta y un gasto actual en educación secundaria (% del gasto total en instituciones públicas secundarias) que varía.

Por otro lado, los Data points 1 y 20 al 37 se han asignado a diferentes clusters (Clusters 0, 2 y 3). Esto indica que estos puntos de datos tienen características distintas en términos de los indicadores seleccionados.

# Selección del número óptimo de clusters

Se implementó un bucle for para probar diferentes valores de número de clusters (desde 2 hasta 10) y se utilizó la métrica de Silhouette Score para evaluar la calidad de los clusters generados. El número óptimo de clusters se seleccionó como aquel que tuviera el mayor valor de Silhouette Score.

```python
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
# ==================================================
# Prepare the data for clustering
# ==================================================
pivot_table = pd.pivot_table(filtered_df, values='Value', index=['#country+name', '#country+code', 'Year'], columns='Indicator Code')
pivot_table.reset_index(inplace=True)
# ==================================================
# Create a list of indicator codes
# ==================================================
indicator_codes = ['UIS.GER.123', 'SE.SEC.NENR', 'SE.PRM.NENR', 'SE.SEC.ENRR.LO', 'SE.XPD.CPRM.ZS', 'SE.XPD.CSEC.ZS']
# ==================================================
# Select the columns corresponding to the indicator codes
# ==================================================
X = pivot_table[indicator_codes].values
# ==================================================
# Impute missing values
# ==================================================
imputer = SimpleImputer(strategy='mean')
X = imputer.fit_transform(X)
# ==================================================
# Normalize the data
# ==================================================
scaler = StandardScaler()
X = scaler.fit_transform(X)
# ==================================================
# Select the optimal number of clusters
# ==================================================
best_score = -1
best_labels = None
best_n_clusters = None
# ==================================================
# Iterate over different values of n_clusters
# ==================================================
for n_clusters in range(2, 10):
    # ==================================================
    # Create an instance of AgglomerativeClustering with current n_clusters
    # ==================================================
    agglomerative = AgglomerativeClustering(n_clusters=n_clusters)
    # ==================================================
    # Fit the data X and obtain cluster labels
    # ==================================================
    labels = agglomerative.fit_predict(X)
    # ==================================================
    # Calculate the silhouette score to evaluate cluster quality
    # ==================================================
    score = silhouette_score(X, labels)
    # ==================================================
    # Check if the current score is better than the best score so far
    # ==================================================
    if score > best_score:
        # ==================================================
        # Update the best score, best labels, and best number of clusters
        # ==================================================
        best_score = score
        best_labels = labels
        best_n_clusters = n_clusters
# ==================================================
# Print the best number of clusters and silhouette score
# ==================================================
print(f"Best number of clusters: {best_n_clusters}")
print(f"Best silhouette score: {best_score}")
# ==================================================
# Print the cluster labels
# ==================================================
for i, label in enumerate(best_labels):
    print(f"Data point {i+1}: Cluster {label}")
# ==================================================
# Visualize the clusters
# ==================================================
plt.scatter(X[:, 0], X[:, 1], c=best_labels)
plt.xlabel('Enrollment Ratios')
plt.ylabel('Education Expenditure')
plt.title('Agglomerative Clustering: Enrollment Ratios vs Education Expenditure')
plt.show()
```

Toma valores en el rango de -1 a 1, donde un valor más cercano a 1 indica una mejor separación entre los clusters.

En este caso, un valor de 0.5315 sugiere que los clusters generados tienen una buena separación y cohesión relativa entre ellos.

Se realiza una búsqueda exhaustiva de los parámetros del algoritmo DBSCAN para encontrar la combinación que produce el mejor resultado. Se exploran diferentes valores de epsilon y min_samples para determinar los límites de densidad y el número mínimo de muestras para formar un cluster. Se utiliza el índice de silueta (silhouette_score) como medida de evaluación para seleccionar los mejores parámetros.

```python
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
# ==================================================
# Prepare the data for clustering
# ==================================================
pivot_table = pd.pivot_table(filtered_df, values='Value', index=['#country+name', '#country+code', 'Year'], columns='Indicator Code')
pivot_table.reset_index(inplace=True)
# ==================================================
# Create a list of indicator codes
# ==================================================
indicator_codes = ['UIS.GER.123', 'SE.SEC.NENR', 'SE.PRM.NENR', 'SE.SEC.ENRR.LO', 'SE.XPD.CPRM.ZS', 'SE.XPD.CSEC.ZS']
# ==================================================
# Select the columns corresponding to the indicator codes
# ==================================================
X = pivot_table[indicator_codes].values
# ==================================================
# Impute missing values
# ==================================================
imputer = SimpleImputer(strategy='mean')
X = imputer.fit_transform(X)
# ==================================================
# Normalize the data
# ==================================================
scaler = StandardScaler()
X = scaler.fit_transform(X)
# ==================================================
# Perform parameter selection
# ==================================================
eps_values = [0.1, 0.5, 1.0]  # Example values for epsilon
min_samples_values = [3, 5, 10]  # Example values for minimum samples

best_score = -1
best_labels = None
best_eps = None
best_min_samples = None
# ==================================================
# Iterate over different values of eps
# ==================================================
for eps in eps_values:
    for min_samples in min_samples_values:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        labels = dbscan.fit_predict(X)
        # ==================================================
        # Check if multiple clusters are formed
        # ==================================================
        if len(set(labels)) > 1:
            # ==================================================
            # Evaluate the clustering using silhouette score
            # ==================================================
            score = silhouette_score(X, labels)
            if score > best_score:
                best_score = score
                best_labels = labels
                best_eps = eps
                best_min_samples = min_samples
# ==================================================
# Print the best parameters and silhouette score
# ==================================================
print(f"Best parameters - Epsilon: {best_eps}, Min samples: {best_min_samples}")
print(f"Best silhouette score: {best_score}")
# ==================================================
# Print the cluster labels
# ==================================================
for i, label in enumerate(best_labels):
    print(f"Data point {i+1}: Cluster {label}")
# ==================================================
# Visualize the clusters
# ==================================================
plt.scatter(X[:, 0], X[:, 1], c=best_labels)
plt.xlabel('Enrollment Ratios')
plt.ylabel('Education Expenditure')
plt.title('DBSCAN Clustering: Enrollment Ratios vs Education Expenditure')
plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
# ==================================================
# Select the variables of interest
# ==================================================
selected_indicators = ['UIS.GER.123', 'SE.SEC.NENR', 'SE.PRM.NENR', 'SE.SEC.ENRR.LO', 'SE.XPD.CPRM.ZS', 'SE.XPD.CSEC.ZS']
filtered_df = df_education[df_education['Indicator Code'].isin(selected_indicators)]
# ==================================================
# Prepare the data for clustering
# ==================================================
pivot_table = pd.pivot_table(filtered_df, values='Value', index=['#country+name', '#country+code', 'Year'], columns='Indicator Code')
pivot_table.reset_index(inplace=True)
# ==================================================
# Impute missing values
# ==================================================
imputer = SimpleImputer(strategy='mean')
X = imputer.fit_transform(pivot_table[selected_indicators])
# ==================================================
# Determine the optimal number of clusters using the elbow method
# ==================================================
wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
# ==================================================
# Print the WCSS values
# ==================================================
print("WCSS values:")
for k, wcss_value in enumerate(wcss):
    print(f"K={k+1}: WCSS={wcss_value:.2f}")
# ==================================================
# Plot the elbow curve
# ==================================================
plt.plot(range(1, 11), wcss)
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.title('Elbow Method for Optimal K')
plt.show()
```

# Evaluating Affinity Propagation Clustering

Cohesión interna:
La cohesión interna mide la compacidad de los clusters, es decir, qué tan cerca están los puntos dentro de cada cluster. Una métrica común para la cohesión interna es la suma de las distancias cuadradas de los puntos dentro de cada cluster a su centroide. Se puede utilizar la función cdist de scipy.spatial.distance para calcular las distancias entre los puntos y los centroides. Luego, se puede sumar estas distancias para obtener la cohesión interna del modelo.

Separación entre clústeres:
La separación entre clústeres mide qué tan bien están separados los clusters entre sí. Una métrica común es la distancia promedio entre los centroides de los clusters. Se puede calcular utilizando la misma matriz de distancias que se obtuvo en el paso anterior.

Coeficiente de silueta:
El coeficiente de silueta es una medida más completa que combina tanto la cohesión interna como la separación entre clústeres. Se puede calcular utilizando la función silhouette_score de sklearn.metrics.

# Evaluating Agglomerative Clustering

```python
from scipy.cluster.hierarchy import linkage, fcluster
from scipy.spatial.distance import cdist
# ==================================================
# Calculate the linkage matrix
# ==================================================
linkage_matrix = linkage(X, method='ward')
# ==================================================
# Get the cluster labels
# ==================================================
labels = fcluster(linkage_matrix, t=2, criterion='maxclust')
# ==================================================
# Calculate Cohesion
# ==================================================
cohesion = 0
for cluster in np.unique(labels):
    cluster_indices = np.where(labels == cluster)[0]
    distances = cdist(X[cluster_indices], X[cluster_indices], metric='euclidean')
    cohesion += np.sum(distances)
print("Cohesion:", cohesion)
# ==================================================
# Calculate Separation
# ==================================================
separation = 0
for i in np.unique(labels):
    cluster_i_indices = np.where(labels == i)[0]
    for j in np.unique(labels):
        if j != i:
            cluster_j_indices = np.where(labels == j)[0]
            distances = cdist(X[cluster_i_indices], X[cluster_j_indices], metric='euclidean')
            separation += np.mean(distances)
separation /= len(np.unique(labels)) * (len(np.unique(labels)) - 1)
print("Separation:", separation)
```

Al analizar los resultados obtenidos del modelo de Agglomerative Clustering, podemos observar lo siguiente:

Cohesión: El valor de cohesión obtenido es de 3067.345185014639. Este valor indica la suma de las distancias entre los puntos dentro de cada cluster. Cuanto menor sea el valor de cohesión, mayor será la cohesión interna de los clusters, lo que implica que los puntos dentro de cada cluster están más cercanos entre sí. En este caso, un valor de cohesión de 3067.345185014639 indica una cohesión relativamente alta.

Separación: El valor de separación obtenido es de 4.722775929786755. Este valor indica la distancia promedio entre los puntos de diferentes clusters. Cuanto mayor sea el valor de separación, mayor será la separación entre los clusters, lo que implica que los puntos de diferentes clusters están más alejados entre sí. En este caso, un valor de separación de 4.722775929786755 indica una separación relativamente alta.

Estos resultados sugieren que el modelo de Agglomerative Clustering ha logrado formar clusters con una cohesión interna alta y una separación adecuada entre los clusters.
