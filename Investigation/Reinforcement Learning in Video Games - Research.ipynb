# 1. Introducción

## ¿Qué es Aprendizaje automático en vídeo juegos (Reinforcement Learning)?

El Aprendizaje Automático en Videojuegos (Reinforcement Learning), es un enfoque de la inteligencia artificial que se utiliza para enseñar a los agentes de IA a jugar y mejorar su desempeño en juegos a través de la interacción con el entorno del juego.

Reinforcement learning resuelve problemas mediante prueba y error. Basándose en datos, entrena distintos escenarios para tomar una serie de decisiones. En función de las acciones realizadas recibe recompensas o penalizaciones (Monroy, 2023)

El aprendizaje por refuerzo consiste en aprender qué hacer, cómo mapear situaciones a acciones, para maximizar una señal numérica de recompensa. No se le dice al aprendiz qué acciones tomar, sino que debe descubrir qué acciones generan la mayor recompensa probándolas. Las acciones pueden afectar no solo la recompensa inmediata, sino también la próxima situación y, a través de eso, todas las recompensas subsecuentes. Estas dos características: búsqueda mediante prueba y error y recompensa demorada, son las dos características distintivas más importantes del aprendizaje por refuerzo (Sutton y Barto, 2020).


A medida que el agente juega repetidamente, utiliza técnicas de aprendizaje automático para ajustar su estrategia y mejorar su desempeño. Esto puede implicar el uso de algoritmos como Q-Learning, Deep Q-Networks (DQN), Policy Gradient, entre otros, que permiten al agente aprender a tomar decisiones informadas y adaptarse a las dinámicas cambiantes del juego.

Se ha utilizado de forma eficaz en una amplia gama de juegos, desde juegos de mesa clásicos hasta juegos de video más complejos. Los agentes de IA entrenados con Reinforcement Learning han logrado resultados impresionantes, incluso superando a jugadores humanos en algunos casos.

Este enfoque tiene aplicaciones tanto en el desarrollo de juegos, donde se pueden crear personajes controlados por IA más inteligentes y desafiantes, como en la investigación en inteligencia artificial, donde se exploran nuevas técnicas y algoritmos para mejorar el aprendizaje automático en entornos de juegos.

## Fundamento Matemático

El Aprendizaje por Reforzamiento tiene un sólido fundamento matemático que se basa en la teoría de control óptimo y la teoría de decisiones secuenciales. A continuación, se describen algunos conceptos matemáticos fundamentales utilizados en el Aprendizaje por Reforzamiento:

### 1. Proceso de decisión de Markov (MDP):

 El MDP es un modelo matemático utilizado para representar un problema de decisión secuencial en el Aprendizaje por Reforzamiento. Consiste en un conjunto de estados, acciones, probabilidades de transición y recompensas asociadas. La propiedad clave del MDP es la propiedad de Markov, que establece que la distribución de probabilidad futura depende únicamente del estado actual y la acción tomada, no de la historia completa.
 
 ### Un MDP está compuesto por:   
- Conjunto de estados S  
- Conjunto de acciones A  
- Función de transición de estados T(s, a, s')  
- Función de recompensa R(s, a, s')

Se compone de un conjunto de estados posibles (S), un conjunto de acciones posibles (A), una función de transición que describe las probabilidades de pasar de un estado a otro dado una acción (T(s, a, s')), y una función de recompensa que asigna recompensas a las transiciones entre estados (R(s, a, s')).

(Sutton y Barto, 2020)
    

  El aprendiz y tomador de decisiones se llama **agente**. Aquello con lo que interactúa, que comprende todo lo que está fuera del agente, se llama **entorno**. Estos interactúan continuamente, con el agente seleccionando **acciones** y el entorno respondiendo a estas acciones y presentando nuevas **situaciones** al agente. El entorno también genera **recompensas**, valores numéricos especiales que el agente busca maximizar con el tiempo a través de su elección de acciones.


### 2. Política:

 Una política en el Aprendizaje por Reforzamiento es una función que mapea estados a acciones. Puede ser determinista, seleccionando una acción específica para cada estado, o estocástica, asignando una distribución de probabilidad sobre las acciones para cada estado.

### Una política π es una función que mapea estados a acciones:  
$π(s) -> a$

Una política (π) es una función que mapea estados a acciones.
Dado un estado (s), la política devuelve una acción (a) que debe ser tomada en ese estado.

### 3. Función de valor: 

La función de valor asigna un valor numérico a cada estado o par estado-acción en un MDP. El valor representa la "calidad" de un estado o acción en términos de la recompensa esperada a largo plazo. Hay dos tipos principales de funciones de valor: la función de valor de estado (V-Value) y la función de valor de acción (Q-Value).

Función de valor de estado (V-Value):
$V(s) = E[R_t+1 + γ * R_t+2 + γ^2 * R_t+3 + ... | S_t = s]$

La función de valor de estado (V-Value) (V(s)) representa la recompensa esperada acumulada a largo plazo comenzando desde un estado dado.

Función de valor de acción (Q-Value):
$Q(s, a) = E[R_t+1 + γ * R_t+2 + γ^2 * R_t+3 + ... | S_t = s, A_t = a]$

La función de valor de acción (Q-Value) (Q(s, a)) representa la recompensa esperada acumulada a largo plazo comenzando desde un estado dado y tomando una acción específica.

Estas funciones de valor son utilizadas para evaluar la "calidad" de los estados o acciones en términos de la recompensa esperada a largo plazo.

### 4. Función de valor óptima: 

La función de valor óptima es la función que asigna los valores máximos a cada estado o par estado-acción en un MDP. Indica la máxima recompensa esperada que se puede obtener siguiendo una política óptima.

Función de valor de estado óptima (V-Value óptima):
$V*(s) = max(Q*(s, a)) para todo a en A$

Función de valor de acción óptima (Q-Value óptima):
$Q*(s, a) = E[R_t+1 + γ * max(Q*(s', a')) | S_t = s, A_t = a]$

La función de valor de estado óptima (V*(s)) representa la recompensa esperada acumulada a largo plazo bajo la mejor política posible para cada estado.
La función de valor de acción óptima (Q*(s, a)) representa la recompensa esperada acumulada a largo plazo bajo la mejor política posible para cada par estado-acción.

### 5.Algoritmos de actualización de valor: 

Los algoritmos de Aprendizaje por Reforzamiento utilizan métodos de actualización iterativa para estimar y actualizar las funciones de valor. Algunos algoritmos comunes incluyen Q-Learning, Sarsa, Deep Q-Networks (DQN) y Métodos de Gradiente de Política.

**Q-Learning:**

Es un algoritmo de Aprendizaje por Reforzamiento que actualiza la función de valor de acción (Q) utilizando la diferencia entre la recompensa actual y la estimación del valor futuro óptimo.

$Q(s, a) ← Q(s, a) + α * (R + γ * max(Q(s', a')) - Q(s, a))$

Donde:

* Q(s, a) es el valor de Q para el estado s y la acción a.
α es la tasa de aprendizaje, que controla cuánto se actualiza el valor de Q en cada iteración.
* r es la recompensa recibida por tomar la acción a en el estado s.
γ es el factor de descuento, que determina la importancia relativa de las recompensas futuras frente a las inmediatas.
* s' es el estado resultante después de tomar la acción a en el estado s.
* a' es la acción óptima en el estado s' según la función Q.
* El algoritmo Q-learning se basa en un proceso de exploración y explotación, donde el agente explora el entorno para descubrir nuevas acciones y actualiza su conocimiento utilizando las recompensas recibidas. Con el tiempo, el agente aprende una política óptima, es decir, una estrategia para elegir las mejores acciones en cada estado.


**Sarsa:**

Es un algoritmo de Aprendizaje por Reforzamiento que actualiza la función de valor de acción (Q) utilizando la diferencia entre la recompensa actual y la estimación del valor futuro según la política actual.

$Q(s, a) ← Q(s, a) + α * (R + γ * Q(s', a') - Q(s, a))$


**Deep Q-Networks (DQN):**

Utiliza una red neuronal profunda para aproximar la función de valor de acción óptima (Q*) y se entrena minimizando el error cuadrático entre las estimaciones y las recompensas reales.

$Loss = (Q(s, a) - (R + γ * max(Q(s', a'))))^2$


**Métodos de Gradiente de Política:**

Utiliza el gradiente de la función de valor con respecto a los parámetros de la política para guiar la actualización de los mismos. En cada paso de optimización, se calcula el gradiente y se realiza un pequeño ajuste a los parámetros de la política en la dirección que maximice la recompensa esperada.

$θ ← θ + α * ∇(log π(a|s)) * Q(s, a)$

## a. Silly Snake game por un humano

Esta versión del juego es la original, está pensada para que la juegue un humano. Incluye un método principal.

```python
# ==========================================================
# Vídeo de guía, referencia en el apartado de Referencias:
# https://youtu.be/--nsd2ZeYvs?t=1256
# ==========================================================
# Se importa las bibliotecas necesarias
# ==========================================================
import pygame
import random
from enum import Enum
from collections import namedtuple
# ==========================================================
# Se crea una clase dirección que sera un ENUM donde tendra
# las direcciones posibles de la serpiente
# ==========================================================
class Direction(Enum):
    RIGHT = 1
    LEFT = 2
    UP = 3
    DOWN = 4
# ==========================================================
# Se establece un el punto con eje X y Y
# ==========================================================
Point = namedtuple("Point", "x, y")
# ==========================================================
# sustituimos pygame.init() por el inicio explícito de iniciar la pantalla y las fuentes
# esto evitará que se inicie el sonido, lo que puede causar problemas en las computadoras portátiles
# ==========================================================
pygame.display.init()
pygame.font.init()
font = pygame.font.Font(pygame.font.get_default_font(), 25)

# rgb colors
# ==========================================================
# Se establece los colores RGB
# ==========================================================
WHITE = (255, 255, 255)
RED = (200,0,0)
BLUE1 = (0, 0, 255)
BLUE2 = (0, 100, 255)
BLACK = (0,0,0)
# ==========================================================
# Se define el tamaño del bloque y velocidad
# ==========================================================
BLOCK_SIZE = 20
SPEED = 10
# ==========================================================
# Se declara la clase SnakeGame
# ==========================================================
class SnakeGame:
    # ==========================================================
    # Se definen el ancho, la altura y el nombre del jugador
    # ==========================================================
    def __init__(self, width=640, height=480, playerName="Player"):
        self.w = width
        self.h = height
        self.playerName = playerName
        # ==========================================================
        # El display inicial
        # ==========================================================
        self.display = pygame.display.set_mode((self.w, self.h))
        pygame.display.set_caption("Silly Snake")
        self.clock = pygame.time.Clock()

        # ==========================================================
        # La inicialización del juego
        # ==========================================================
        self.direction = Direction.RIGHT
        # ==========================================================
        # Punto de partida en el medio
        # ==========================================================
        self.head = Point(self.w / 2, self.h / 2)
        # ==========================================================
        # snake estará definido como la cabeza y se establece el 
        # crecimiento que debe de tener en caso que coma una "fruta"
        # ==========================================================
        self.snake = [self.head,
                      Point(self.head.x - BLOCK_SIZE, self.head.y),
                      Point(self.head.x - (2 * BLOCK_SIZE), self.head.y)]
        # ==========================================================
        # Se inicializa los diferentes scores
        # ==========================================================
        self.score = 0
        self.food = None
        self.placeFood()
    # ==========================================================
    # PlaceFood se encargará de establecer si colocar la fruta 
    # que debe de comer la serpiente de manera aleatoria en el 
    # tablero con una coordenada x y aleatoria
    # ==========================================================
    def placeFood(self):
        x = random.randint(0, (self.w-BLOCK_SIZE )//BLOCK_SIZE )*BLOCK_SIZE
        y = random.randint(0, (self.h-BLOCK_SIZE )//BLOCK_SIZE )*BLOCK_SIZE
        # ==========================================================
        # Se define el punto
        # ==========================================================
        self.food = Point(x, y)
        # ==========================================================
        # En el caso que la fruta "caiga" encima de la serpiente se
        # deberá de colocar la fruta en otro lugar
        # ==========================================================
        if self.food in self.snake:
            self.placeFood()
    # ==========================================================
    # PlayStep es basicamente el paso de debe tener el juego
    # ==========================================================
    def playStep(self):
        # ==========================================================
        # Se colectan las entradas del "usuario"
        # ==========================================================
        for event in pygame.event.get():
            # ==========================================================
            # En el caso que el evento sea QUIT se saldrá del juego
            # ==========================================================
            if event == pygame.QUIT:
                pygame.quit()
                quit()
            # ==========================================================
            # Los posibles movimientos que tiene pygame hay que adaptarlos
            # a los movimientos de la serpiente
            # LEFT, RIGHT, UP y DOWN
            # ==========================================================
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_LEFT:
                    self.direction = Direction.LEFT
                elif event.key == pygame.K_RIGHT:
                    self.direction = Direction.RIGHT
                elif event.key == pygame.K_UP:
                    self.direction = Direction.UP
                elif event.key == pygame.K_DOWN:
                    self.direction = Direction.DOWN

        # ==========================================================
        # Se mueve serpiente se movera a direcciones aleatorias
        # si es el caso que no "caiga" en algun caso especial
        # ==========================================================
        self.moveSnake(self.direction)
        # 3. check if game over
        # ==========================================================
        # Verifica si hay una choque, en dado caso gameOver = True
        # ==========================================================
        if self.isCollision():
            gameOver = True
            return gameOver, self.score
        
        # 4. move snake or place food
        # ==========================================================
        # En el caso que la fruta es comida por la cabeza el score aumente
        # y se llama PlaceFood para colocar una nueva fruta
        # ==========================================================
        if self.head == self.food:
            self.score += 1
            self.placeFood()
        else:
            # ==========================================================
            # En dado caso se hace snake.pop()
            # ==========================================================
            self.snake.pop()

        # ==========================================================
        # Se actualiza la pantalla del juego
        # ==========================================================
        self.updateUi()
        self.clock.tick(SPEED)
        # ==========================================================
        # En caso que no haya un choque gameOver = False
        # ==========================================================
        gameOver = False

        return gameOver, self.score
    
    # ==========================================================
    # En el caso que haya una colision
    # ==========================================================
    def isCollision(self):
        # ==========================================================
        # En el caso de que choque en el borde
        # ==========================================================
        if self.head.x > self.w - BLOCK_SIZE or self.head.x < 0:
            return True
        if self.head.y > self.h - BLOCK_SIZE or self.head.y < 0:
            return True

        # ==========================================================
        # En el caso que choque con ella misma
        # ==========================================================
        if self.head in self.snake[1:]:
            return True
        return False

    # ==========================================================
    # Par mover la serpiente, solo se necesita la direccion como
    # parametro
    # ==========================================================
    def moveSnake(self, direction: Direction):
        # ==========================================================
        # Coordenadas
        # ==========================================================
        x = self.head.x
        y = self.head.y
        # ==========================================================
        # Movimientos posibles de la serpiente
        # ==========================================================
        if direction == Direction.RIGHT:
            x += BLOCK_SIZE
        elif direction == Direction.LEFT:
            x -= BLOCK_SIZE
        elif direction == Direction.DOWN:
            y += BLOCK_SIZE
        elif direction == Direction.UP:
            y -= BLOCK_SIZE
        self.head = Point(x, y)
        # ==========================================================
        # En el caso que coma una fruta aumenta de tamaño
        # ==========================================================
        self.snake.insert(0, self.head)
        
    # ==========================================================
    # Se actualiza la UI de la pantalla
    # ==========================================================
    def updateUi(self):
        self.display.fill(BLACK)
        # ==========================================================
        # Se pinta el largo de la serpiente
        # ==========================================================
        for p in self.snake:
            pygame.draw.rect(self.display, BLUE1, pygame.Rect(p.x, p.y, BLOCK_SIZE, BLOCK_SIZE))
            pygame.draw.rect(self.display, BLUE2, pygame.Rect(p.x + 4, p.y + 4, 12, 12)) # p.x + 4, p.y + 4, 12, 12
                                                                                         # coordenada x +  coordenada y + ancho + altura
        # ==========================================================
        # Se pinta la fruta que tiene que comer la serpiente
        # ==========================================================
        pygame.draw.rect(self.display, RED, pygame.Rect(self.food.x, self.food.y, BLOCK_SIZE, BLOCK_SIZE))
        # ==========================================================
        # Se actualiza los valores de pantalla como el score, speed, player
        # ==========================================================
        scoreText = font.render("Score: " + str(self.score) + " Speed: " + str(SPEED) + " Player: " + self.playerName, True, WHITE)
        self.display.blit(scoreText, [0, 0])
        pygame.display.flip()

# ==========================================================
# Se declara PlayGame
# ==========================================================
def playGame(playerName = "Stefan"):
    # ==========================================================
    # Se llama el juego principal
    # ==========================================================
    game = SnakeGame(playerName=playerName)
    # ==========================================================
    # Bucle para que el juego
    # ==========================================================
    while True:
        gameOver, score = game.playStep()
        # ==========================================================
        # En el caso gameOver == True se termina el juego (reinicio)
        # ==========================================================
        if gameOver == True:
            break
    # ==========================================================
    # Se imprime el score final
    # ==========================================================
    print("Game Over, final score: ", score)
    pygame.quit()

# include a main function, to be able to run the script, for notebooks we can comment this out
if __name__ == "__main__":
    playGame()
```

## b. Se define el ambiente

El **Ambiente** es el "place" donde el agente aprenderá y realizará acciones.

En nuestro caso, el **Entorno** es una versión del juego Silly Snake adaptada para ser jugable por la IA.

```python
# ==========================================================
# Se importan las bibliotecas necesarias
# ==========================================================
import pygame
import random
from enum import Enum
from collections import namedtuple
import numpy as np
# ==========================================================
# Dirección 
# ==========================================================
class Direction(Enum):
    RIGHT = 1
    LEFT = 2
    UP = 3
    DOWN = 4

Point = namedtuple("Point", "x, y")

# ==========================================================
# Sustituimos pygame.init() por el inicio explícito de iniciar la pantalla y las fuentes
# esto evitará que se inicie el sonido, lo que puede causar problemas en las computadoras portátiles
# ==========================================================
pygame.display.init()
pygame.font.init()
font = pygame.font.Font(pygame.font.get_default_font(), 25)

# ==========================================================
# Se definen los colores RGB
# ==========================================================
WHITE = (255, 255, 255)
RED = (200,0,0)
BLUE1 = (0, 0, 255)
BLUE2 = (0, 100, 255)
BLACK = (0,0,0)
# ==========================================================
# Se declara el tamaño del bloque y la velocidad
# ==========================================================
BLOCK_SIZE = 20
SPEED = 120

# ==========================================================
# Clase de la serpiente con IA
# ==========================================================
class SillySnakeGameAi:
    # ==========================================================
    # Se declara el ancho, el alto y el jugador
    # ==========================================================
    def __init__(self, width=640, height=480, playerName="Player"):
        self.w = width
        self.h = height
        self.playerName = playerName
        # ==========================================================
        # El display inicial
        # ==========================================================
        self.display = pygame.display.set_mode((self.w, self.h))
        pygame.display.set_caption("Silly Snake")
        self.clock = pygame.time.Clock()

        # ==========================================================
        # Se reseta el juego
        # ==========================================================
        self.reset()
        
    # ==========================================================
    # El reset consite en colocar la serpiente en el centro de la 
    # pantalla cada vez que piede el intento
    # ==========================================================
    def reset(self):
        # ==========================================================
        # Donde se mueve la serpiente
        # ==========================================================
        self.direction = Direction.RIGHT
        # ==========================================================
        # Se coloca la serpiente en el medio de la pantalla
        # ==========================================================
        self.head = Point(self.w / 2, self.h / 2)
        # ==========================================================
        # Se declara el tamaño de la serpiente
        # ==========================================================
        self.snake = [self.head,
                      Point(self.head.x - BLOCK_SIZE, self.head.y),
                      Point(self.head.x - (2 * BLOCK_SIZE), self.head.y)]
        # ==========================================================
        # Se declara diferentes scores
        # ==========================================================
        self.score = 0
        self.food = None
        # ==========================================================
        # Se remplaza el lugar de la fruta
        # ==========================================================
        self.placeFood()
        # ==========================================================
        # La iteración
        # ==========================================================
        self.frameIteration = 0

    # ==========================================================
    # Colocar la fruta en un nuevo sitio
    # ==========================================================
    def placeFood(self):
        # ==========================================================
        # Una coordenada x y aleatoria para colocar la fruta
        # ==========================================================
        x = random.randint(0, (self.w-BLOCK_SIZE )//BLOCK_SIZE )*BLOCK_SIZE
        y = random.randint(0, (self.h-BLOCK_SIZE )//BLOCK_SIZE )*BLOCK_SIZE
        self.food = Point(x, y)
        # ==========================================================
        # En el caso que la fruta caiga en la serpiete debe de generar
        # la fruta en otro sitio
        # ==========================================================
        if self.food in self.snake:
            self.placeFood()
    # ==========================================================
    # PlayStep es basicamente el paso de debe tener el juego
    # ==========================================================
    def playStep(self, action):
        # ==========================================================
        # Se incrementa frame iteration
        # ==========================================================
        self.frameIteration += 1
        # ==========================================================
        # En el caso que el evento sea QUIT sale del juego para luego
        # para luego este mismo reinicie en otra etapa
        # ==========================================================
        for event in pygame.event.get():
            if event == pygame.QUIT:
                pygame.quit()
                quit()

        # ==========================================================
        # Se mueve la serpiente despues de haber colocado la fruta
        # ==========================================================
        self.moveSnake(action)

        # ==========================================================
        # El contador de la recompensa
        # ==========================================================
        reward = 0

        # ==========================================================
        # Revisar si el juego ha acabado
        # ==========================================================
        if self.isCollision() or self.frameIteration > 100 * len(self.snake):
            gameOver = True
            reward -= 10
            return reward, gameOver, self.score

        # ==========================================================
        # En el caso que la fruta sea comida por la cabeza incremeta
        # el tamaño el score
        # ==========================================================
        if self.head == self.food:
            self.score += 1
            reward = 10
            # ==========================================================
            # Se remplaza el lugar de la fruta
            # ==========================================================
            self.placeFood()
        else:
            self.snake.pop()

        # ==========================================================
        # Se actualiza la UI
        # ==========================================================
        self.updateUi()
        self.clock.tick(SPEED)
        # ==========================================================
        # gameOver = False en caso que no haya caigo en ningun otra 
        # condicional
        # ==========================================================
        gameOver = False
        return reward, gameOver, self.score
    
    # ==========================================================
    # En el caso que haya un choque
    # ==========================================================
    def isCollision(self, p: Point = None):
        if p == None:
            p = self.head
        # ==========================================================
        # En el caso que choque en el borde
        # ==========================================================
        if p.x > self.w - BLOCK_SIZE or p.x < 0:
            return True
        if p.y > self.h - BLOCK_SIZE or p.y < 0:
            return True

        # ==========================================================
        # En el caso que choque con el cuerpo de la serpiente
        # ==========================================================
        if p in self.snake[1:]:
            return True
        return False
    # ==========================================================
    # MoveSnake se encarga de mover a la serpiente:
    #       action: action -> [straigth, right, left]
    # ==========================================================
    def moveSnake(self, action):
        # action -> [straigth, right, left]
        # ==========================================================
        # Direcciones posibles
        # ==========================================================
        clockWiseDirections = [Direction.RIGHT, Direction.DOWN, Direction.LEFT, Direction.UP]
        # ==========================================================
        # Indice de la direccion actual
        # ==========================================================
        currentDirectionIndex = clockWiseDirections.index(self.direction)
        # ==========================================================
        # Nueva direccion
        # ==========================================================
        newDirection = self.direction
        # ==========================================================
        # Se establece una nueva dirección en base a la dirección actual
        # ==========================================================
        if np.array_equal(action, [0, 1, 0]):
            newDirection = clockWiseDirections[(currentDirectionIndex + 1) % 4]
        # ==========================================================
        # Se establece una nueva dirección en base a la dirección actual
        # ==========================================================
        elif np.array_equal(action, [0, 1, 0]):
            newDirection = clockWiseDirections[(currentDirectionIndex - 1) % 4]
        # ==========================================================
        # Se establece la nueva dirección
        # ==========================================================
        self.direction = newDirection
        # ==========================================================
        # Coordenadas x y
        # ==========================================================
        x = self.head.x
        y = self.head.y
        # ==========================================================
        # Direcciones posibles de la serpiente con IA
        # ==========================================================
        if self.direction == Direction.RIGHT:
            x += BLOCK_SIZE
        elif self.direction == Direction.LEFT:
            x -= BLOCK_SIZE
        elif self.direction == Direction.DOWN:
            y += BLOCK_SIZE
        elif self.direction == Direction.UP:
            y -= BLOCK_SIZE
        self.head = Point(x, y)
        # ==========================================================
        # En caso para que la serpiente crezca
        # ==========================================================
        self.snake.insert(0, self.head)

    # ==========================================================
    # Funcion para actualizar la UI
    # ==========================================================
    def updateUi(self):
        # ==========================================================
        # Fondo NEGRO
        # ==========================================================
        self.display.fill(BLACK)
        # ==========================================================
        # Poder pintar la serpiente segun su largo
        # ==========================================================
        for p in self.snake:
            pygame.draw.rect(self.display, BLUE1, pygame.Rect(p.x, p.y, BLOCK_SIZE, BLOCK_SIZE))
            pygame.draw.rect(self.display, BLUE2, pygame.Rect(p.x + 4, p.y + 4, 12, 12))
        # ==========================================================
        # Se pinta la fruta
        # ==========================================================
        pygame.draw.rect(self.display, RED, pygame.Rect(self.food.x, self.food.y, BLOCK_SIZE, BLOCK_SIZE))
        # ==========================================================
        # Se establece una nueva dirección en base a la dirección actual
        # ==========================================================
        scoreText = font.render("Score: " + str(self.score) + " Speed: " + str(SPEED) + " Player: " + self.playerName, True, WHITE)

        self.display.blit(scoreText, [0, 0])

        pygame.display.flip()

    # ==========================================================
    # Se setea el nombre del jugador
    # ==========================================================
    def setPlayerName(self, name):
        self.playerName = name
```

## c.Se define el modelo

Esta es una red neuronal simple utilizada para el entrenamiento, debe tener 11 neuronas de entrada (el tamaño de nuestra matriz de estado) y 3 neuronas de salida, el número de nuestros movimientos potenciales (recto, izquierda derecha).

```python
# ==========================================================
# Bibliotecas necesarias para el modelo
# ==========================================================
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
# ==========================================================
# Se declara la clase LinearQNet
# ==========================================================
class LinearQNet(nn.Module):
    # ==========================================================
    # Para definir el modelo se nesita inputSize, hiddenSize y outputSize
    # ==========================================================
    def __init__(self, inputSize, hiddenSize, outputSize):
        super().__init__()
        # ==========================================================
        # Capa lineal 1
        # ==========================================================
        self.linear1 = nn.Linear(inputSize, hiddenSize)
        # ==========================================================
        # Capa lineal 2
        # ==========================================================
        self.linear2 = nn.Linear(hiddenSize, outputSize)
    # ==========================================================
    # Forward se encarga del cálculo hacia adelante en la red
    # ==========================================================
    def forward(self, X):
        # ==========================================================
        # Aplicar la primera capa lineal
        # ==========================================================
        out = self.linear1(X)
        # ==========================================================
        # Aplicar la función de activación ReLU
        # ==========================================================
        out = F.relu(out)
        # ==========================================================
        # Aplicar la segunda capa lineal
        # ==========================================================
        out = self.linear2(out)
        return out
```

## d. El Entrenador
Esto se utilizará para entrenar nuestro modelo en función de los estados, las acciones y las recompensas proporcionadas por el agente.

```python
# ==========================================================
# Se importa bibliotecas necesarias
# ==========================================================
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
# ==========================================================
# Clase de entrenamiento
# ==========================================================
class QTrainner:
    def __init__(self, model, lr, gamma):
        # ==========================================================
        # Modelo de red neuronal
        # ==========================================================
        self.model = model
        # ==========================================================
        # Tasa de aprendizaje
        # ==========================================================
        self.lr = lr
        # ==========================================================
        # Factor de descuento para las recompensas futuras
        # ==========================================================
        self.gamma = gamma
        # ==========================================================
        # Optimizador Adam para actualizar los pesos del modelo
        # ==========================================================
        self.optimizer = optim.Adam(model.parameters(), self.lr)
        # ==========================================================
        # Función de pérdida de error cuadrático medio (MSELoss)
        # ==========================================================
        self.lossFunction = nn.MSELoss()

    def trainStep(self, state, action, reward, newState, done):
        # ==========================================================
        # Convertir el estado actual en un tensor de PyTorch
        # ==========================================================
        stateTensor = torch.tensor(state, dtype=torch.float)
        # ==========================================================
        # Convertir la acción en un tensor de PyTorch
        # ==========================================================
        actionTensor = torch.tensor(action, dtype=torch.long)
        # ==========================================================
        # Convertir la recompensa en un tensor de PyTorch
        # ==========================================================
        rewardTensor = torch.tensor(reward, dtype=torch.float)
        # ==========================================================
        # Convertir el nuevo estado en un tensor de PyTorch
        # ==========================================================
        newStateTensor = torch.tensor(newState, dtype=torch.float)

        if len(stateTensor.shape) == 1:
            # ==========================================================
            # Añadir una dimensión extra si el estado es un vector
            # ==========================================================
            stateTensor = torch.unsqueeze(stateTensor, 0)
            # ==========================================================
            # Añadir una dimensión extra si el nuevo estado es un vector
            # ==========================================================
            newStateTensor = torch.unsqueeze(newStateTensor, 0)
            # ==========================================================
            # Añadir una dimensión extra si la acción es un escalar
            # ==========================================================
            actionTensor = torch.unsqueeze(actionTensor, 0)
            # ==========================================================
            # Añadir una dimensión extra si la recompensa es un escalar
            # ==========================================================
            rewardTensor = torch.unsqueeze(rewardTensor, 0)
            # ==========================================================
            # Convertir el indicador "done" en una tupla
            # ==========================================================
            done = (done, )

        # ==========================================================
        # Calcular los valores Q predichos para el estado actual
        # ==========================================================
        prediction = self.model(stateTensor)

        # Q_new = 2. reward + gamma * max(next predicted q value) -> only do this if not done
        # ==========================================================
        # Crear una copia de los valores Q predichos
        # ==========================================================
        target = prediction.clone()

        for i in range(len(done)):
            Qnew = rewardTensor[i]
            # ==========================================================
            # Calcular el nuevo valor Q
            # ==========================================================
            if not done[i]:
                Qnew = rewardTensor[i] + self.gamma * torch.max(self.model(newStateTensor[i]))
            # ==========================================================
            # Actualizar el valor Q correspondiente a la acción realizada
            # ==========================================================
            target[i][torch.argmax(actionTensor).item()] = Qnew

        # ==========================================================
        # Reiniciar los gradientes acumulados en el optimizador
        # ==========================================================
        self.optimizer.zero_grad()
        # ==========================================================
        # Calcular la pérdida entre los valores Q predichos y los valores Q objetivo
        # ==========================================================
        loss = self.lossFunction(target, prediction)
        # ==========================================================
        # Realizar la propagación hacia atrás (backpropagation) para calcular los gradientes
        # ==========================================================
        loss.backward()
        # ==========================================================
        # Actualizar los pesos del modelo basado en los gradientes calculados
        # ==========================================================
        self.optimizer.step()
```

## e. Definición del Agente
El **agente** es el componente que hace el aprendizaje y realiza las acciones.

```python
# ==========================================================
# Se importa bibliotecas necesarias
# ==========================================================
import torch
import random
import numpy as np
from collections import deque
# ==========================================================
# Tamaño máximo de la memoria de replay
# ==========================================================
MAX_MEMORY = 100_000
# ==========================================================
# Tamaño del lote de muestras para el entrenamiento
# ==========================================================
BATCH_SIZE = 1000
# ==========================================================
# Tasa de aprendizaje
# ==========================================================
LR = 0.001

class Agent:
    def __init__(self):
        # ==========================================================
        # Número de juegos jugados
        # ==========================================================
        self.numberOfGames = 0
        # controls randomness
        # ==========================================================
        # Controla la aleatoriedad (tasa de exploración)
        # ==========================================================
        self.epsilon = 0
        # discount rate
        # ==========================================================
        # Tasa de descuento para las recompensas futuras
        # ==========================================================
        self.gamma = 0
        # ==========================================================
        # Memoria de replay para almacenar transiciones de juego
        # ==========================================================
        self.memory = deque(maxlen=MAX_MEMORY)

        # the modelneeds to have 11 inputs, as our world status grid has 11 elements and the output is 3, as we have 3 directions
        # ==========================================================
        # Modelo de red neuronal (LinearQNet) con 11 entradas y 3 salidas
        # ==========================================================
        self.model = LinearQNet(11, 256, 3)
        # ==========================================================
        # Entrenador (QTrainner) para el modelo
        # ==========================================================
        self.trainner = QTrainner(self.model, lr=LR, gamma=self.gamma)
    # ==========================================================
    # La funcion getState obtiene el estado del juego
    # ==========================================================
    def getState(self, game):
        # ==========================================================
        # Obtener el estado actual del juego (representación del mundo)
        # ==========================================================
        head = game.head
        # ==========================================================
        # Se definen los puntos LEFT, RIGHT, UP y DOWN
        # ==========================================================
        point_left = Point(head.x - BLOCK_SIZE, head.y)
        point_right = Point(head.x + BLOCK_SIZE, head.y)
        point_up = Point(head.x, head.y - BLOCK_SIZE)
        point_down = Point(head.x, head.y + BLOCK_SIZE)
        # ==========================================================
        # Se definen las direcciones que debe de tener el juego
        # LEFT, RIGHT, UP, DOWN
        # ==========================================================
        direction_left = game.direction == Direction.LEFT
        direction_right = game.direction == Direction.RIGHT
        direction_up = game.direction == Direction.UP
        direction_down = game.direction == Direction.DOWN

        state = [
            # ==========================================================
            # Peligro en línea recta
            # ==========================================================
            (direction_right and game.isCollision(point_right)) or
            (direction_left and game.isCollision(point_left)) or
            (direction_up and game.isCollision(point_up)) or
            (direction_down and game.isCollision(point_down)),

            # ==========================================================
            # Peligro a la derecha
            # ==========================================================
            (direction_up and game.isCollision(point_right)) or
            (direction_down and game.isCollision(point_left)) or
            (direction_left and game.isCollision(point_up)) or
            (direction_right and game.isCollision(point_down)),

            # ==========================================================
            # Peligro a la izquierda
            # ==========================================================
            (direction_down and game.isCollision(point_right)) or
            (direction_up and game.isCollision(point_left)) or
            (direction_right and game.isCollision(point_up)) or
            (direction_left and game.isCollision(point_down)),

            # ==========================================================
            # Dirección de movimiento
            # ==========================================================
            direction_left,
            direction_right,
            direction_up,
            direction_down,

            # ==========================================================
            # Ubicación de la comida
            # ==========================================================
            game.food.x < game.head.x,  # food left
            game.food.x > game.head.x,  # food right
            game.food.y < game.head.y,  # food up
            game.food.y > game.head.y  # food down
        ]

        return np.array(state, dtype=int)
    # ==========================================================
    # remember: Almanena información del juego en base a sus parametros
    #       state: estado del juego
    #       action: accion que se tomó en la jugada
    #       nextState: siguiente estado
    #       done: terminar
    # ==========================================================
    def remember(self, state, action, reward, nextState, done):
        # ==========================================================
        # Almacena la transición del juego en la memoria de replay
        # ==========================================================
        self.memory.append((state, action, reward, nextState, done))

    def trainLongMemory(self):
        # ==========================================================
        # Entrena el modelo utilizando una muestra de la memoria de replay (largo plazo)
        # ==========================================================
        if len(self.memory) < BATCH_SIZE:
            sample = self.memory
        else:
            sample = random.sample(self.memory, BATCH_SIZE)
        # ==========================================================
        # Desempaqueta la muestra en estados, acciones, recompensas, nuevos estados y indicadores de finalización
        # ==========================================================
        states, actions, rewards, nextStates, dones = zip(*sample)
        # ==========================================================
        # Realiza un paso de entrenamiento utilizando la muestra
        # ==========================================================
        self.trainner.trainStep(states, actions, rewards, nextStates, dones)

    def trainShortMemory(self, state, action, reward, nextState, done):
        # ==========================================================
        # Entrena el modelo utilizando una sola transición del juego (corto plazo)
        # ==========================================================
        self.trainner.trainStep(state, action, reward, nextState, done)
    
    def getAction(self, state):
        # ==========================================================
        # Actualiza la tasa de exploración (tradeoff entre exploración y explotación)
        # ==========================================================
        self.epsilon = 80 - self.numberOfGames
        # ==========================================================
        # Acción final a tomar
        # ==========================================================
        finalMove = [0, 0, 0]
        # ==========================================================
        # Toma una acción aleatoria con probabilidad epsilon
        # ==========================================================
        if random.randint(0, 200) < self.epsilon:
            move = random.randint(0, 2)
            finalMove[move] = 1
        else:
            # ==========================================================
            # Convierte el estado en un tensor de PyTorch
            # ==========================================================
            stateTensor = torch.tensor(state, dtype=torch.float)
            # ==========================================================
            # Realiza una predicción utilizando el modelo
            # ==========================================================
            prediction = self.model(stateTensor)
            # ==========================================================
            # Obtiene la acción con el mayor valor Q dada la prediccion
            # ==========================================================
            move = torch.argmax(prediction).item()
            finalMove[move] = 1
        # ==========================================================
        # Devuelve la acción final a tomar
        # ==========================================================
        return finalMove
```

## f. Entrenando y dejando que la IA juegue el juego

Ahora, para ponerlo todo junto, deje que el agente entrene jugando el juego. Como estamos en una libreta no veremos la ventana del juego, solo los logs y gráficas de progreso, pero agregaremos algún GIF para poder ver los resultados.

```python
%%time
# ==========================================================
# Lista para almacenar los puntajes de cada juego
# ==========================================================
scoresHistory = []
# ==========================================================
# Lista para almacenar los puntajes promedio
# ==========================================================
meanScores = []

def train():
    # ==========================================================
    # Puntaje total acumulado
    # ==========================================================
    totalScore = 0
    # ==========================================================
    # Mejor puntaje obtenido
    # ==========================================================
    bestScore = 0
    # ==========================================================
    # Instancia del agente (Agente RL)
    # ==========================================================
    agent = Agent()
    # ==========================================================
    # Instancia del juego (SillySnakeGameAi)
    # ==========================================================
    game = SillySnakeGameAi()

    # ==========================================================
    # Entrenamos el modelo durante 200 juegos
    # ==========================================================
    while agent.numberOfGames < 200:
        game.setPlayerName("Machine epoch " + str(agent.numberOfGames))
        # ==========================================================
        # Obtener el estado anterior
        # ==========================================================
        oldState = agent.getState(game)
        # ==========================================================
        # Realizar movimiento
        # ==========================================================
        finalMove = agent.getAction(oldState)
        # ==========================================================
        # Realizar movimiento y obtener nuevo estado
        # ==========================================================
        reward, done, score = game.playStep(finalMove)
        newState = agent.getState(game)
        # ==========================================================
        # Entrenar memoria corta
        # ==========================================================
        agent.trainShortMemory(oldState, finalMove, reward, newState, done)

        # ==========================================================
        # El agente recuerda aspectos importantes de cada jugada como:
        # oldState, finalMove, reward, newState, done
        # ==========================================================
        agent.remember(oldState, finalMove, reward, newState, done)
        # ==========================================================
        # En el caso que sea done
        # ==========================================================
        if done:
            # ==========================================================
            # Entrenar memoria larga
            # ==========================================================
            game.reset()
            # ==========================================================
            # Incrementa el numero de juegos
            # ==========================================================
            agent.numberOfGames += 1
            # ==========================================================
            # Entrenar memoria larga
            # ==========================================================
            agent.trainLongMemory()
            # ==========================================================
            # Se remplaza score en el caso que bestScore sea mayor
            # ==========================================================
            if score > bestScore:
                bestScore = score
            # ==========================================================
            # El total de scores
            # ==========================================================
            totalScore += score
            # ==========================================================
            # meanScore sera el total de scores entre el numero de juegos
            # ==========================================================
            meanScore = (totalScore / agent.numberOfGames)
            # ==========================================================
            # historial de puntajes
            # ==========================================================
            scoresHistory.append(score)
            meanScores.append(meanScore)
            # ==========================================================
            # Impresion del estado del juego por medio de los siguientes apartados:
            # Game number, Score, Best Score y Mean Scores
            # ==========================================================
            if score == bestScore or agent.numberOfGames % 10 == 0:
                print("Game number: ", agent.numberOfGames, "Score: ", score, "Best Score: ", bestScore, "Mean scores: ", meanScore)

# ==========================================================
# Ejecutar el método de entrenamiento
# ==========================================================
train()
```

Así es como se ve cuando la IA juega, funciona bastante bien. La velocidad está configurada alta, pero esto no molesta a la IA :)

Se han realizado esfuerzos para entrenar agentes de IA capaces de competir contra jugadores humanos en este juego. Por ejemplo, AlphaStar, desarrollado por DeepMind, ha logrado niveles de habilidad comparables a los mejores jugadores humanos en StarCraft II. El aprendizaje por refuerzo este permite que la máquina aprenda a alcanzar una desempeño excelente en juegos a fuerza de jugar millones de partidas contra sí misma, en lugar de aprender del análisis de partidas jugadas por humanos, como hiciera el primer AlphaGo. El objetivo de la IA es maximizar la recompensa esperada, y toma nota de las tácticas que mejor contribuyen a dicho objetivo, para replicarlas después.

# 8. Alance y Limitaciones de Reinforcement Learning

## **Alcance**

- Permite aprender a tomar decisiones en entornos interactivos.
- Se basa en la retroalimentación de recompensas para mejorar las estrategias de toma de decisiones a lo largo del tiempo.
- Es aplicable a una amplia gama de problemas, desde juegos simples hasta problemas complejos en entornos dinámicos.

## **Limitaciones**

- No aborda directamente la construcción, cambio o aprendizaje de la representación del estado.
- Otros métodos que buscan soluciones estáticas, pueden ser más efectivos en ciertos casos.
- Maldición de la dimensionalidad: A medida que aumenta el número de variables y dimensiones en el espacio de estados, el tamaño del espacio de búsqueda aumenta exponencialmente, lo que dificulta el aprendizaje y la generalización de políticas efectivas.
- Diseñar un sistema de recompensas adecuado puede ser complicado. Es importante definir recompensas que guíen al agente hacia el comportamiento deseado, pero esto puede requerir un conocimiento experto y un análisis cuidadoso del problema.
